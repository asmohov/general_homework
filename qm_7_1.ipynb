```python
import numpy as np
from matplotlib import pylab as plt

#define constants
c = 3e8
p = 2.72e-10/c #convert MeV to joules, going to all be in SI
v0 = 3.36e-12 #in joules as before
r0 = 2.6e-15
m = 1.67e-27
hbar = 1.05e-34
k = p/hbar
#define function to return differential cross section
def dsigma_scattering(theta):
    q=2*k*np.sin(theta/2)
    f= 2*m*v0/(q*hbar*hbar)*(np.sin(q*r0)-q*r0*np.cos(q*r0))/(q*q)
    dsigma = f*f
    return dsigma

#plotting section
theta_arr = np.linspace(0,np.pi,100)
y_arr = dsigma_scattering(theta_arr)
plt.semilogy(theta_arr,y_arr)
plt.ylabel('log $\\partial\\sigma$')
plt.xlabel('$\\theta$ in radians')
```
