```python
from matplotlib import pylab as plt
import numpy as np
from scipy import optimize as opt
#App 2B of Valley

#U 235 and 206 Pb
x_vals = np.array([.550,.544,.547,.283,.285,.292])
#To get denominator of 207Pb
x_vals = 1/x_vals

y_vals = np.array([74.6,72.2,73.7,27.5,21.4,24.2])
y_vals = 1/y_vals
# Define solution shape
def func(x, a, b, c):
    return a * np.exp(-b * x) + c

#Fit data to curve
popt,pcurve = opt.curve_fit(func,x_vals,y_vals)
print(popt)

x_fit = np.linspace(1.5,5,1000)
y_fit = func(x_fit,popt[0],popt[1],popt[2])
plt.plot(x_vals,y_vals,color = 'r')
plt.plot(x_fit,y_fit,color = 'g')
plt.show()
```
