```python
import numpy as np
from matplotlib import pylab as plt

#For modal melting
def modalmelt(D,F):
    return (1-F)**((1/D)-1)

D_list = [.001,.01,.5,2,10]

for D in D_list:
    F_list = np.linspace(0.001,.99,1000)
    C_list = []
    for F in F_list:
        C_list.append(modalmelt(D,F))
    plt.semilogy(F_list,C_list,label = str(D))
plt.legend()
plt.ylim([0.0001,10])
plt.xlabel('F')
plt.ylabel('log Cs/Co')
```

```python
#non modal melting, fiducially choosing D = .5

def nonmodalmelt(D,F,P):    
    return ((1-(F*P/D))**(1/P))/(1-F)

D = .5
P_list = [.1,.3,.5,.7,.9]
for P in P_list:
    F_list = np.linspace(0.01,.99,100)
    C_list = []
    for F in F_list:
        C_list.append(nonmodalmelt(D,F,P))
    plt.semilogy(F_list,C_list,label = str(P))
    plt.legend()
plt.xlabel('F')
plt.ylabel('log Cs/Co')
```
